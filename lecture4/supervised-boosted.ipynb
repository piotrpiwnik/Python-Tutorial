```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set()
```

## Data

We use the well-known [handwritten digit recognition](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html) data.

```python
from sklearn.datasets import load_digits

data = load_digits()
```

Let's print some examples:

```python
# from https://jakevdp.github.io/PythonDataScienceHandbook/05.08-random-forests.html#Ensembles-of-Estimators:-Random-Forests

# set up the figure
fig = plt.figure(figsize=(6, 6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits: each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(data.images[i], cmap=plt.cm.binary, interpolation='nearest')
    
    # label the image with the target value
    ax.text(0, 7, str(data.target[i]))
```

#### Training & Test Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target,
                                                    random_state=9462956,
                                                    train_size=0.7)

print(f'   Train   Test   Total')
print(f'X: {X_train.shape[0]}  + {X_test.shape[0]}  = {data.data.shape[0]}')
print(f'y: {y_train.shape[0]}  + {y_test.shape[0]}  = {data.target.shape[0]}')
```

## AdaBoost

There are, in fact, many implementations of boosting meta estimators (meta estimators are those that improve other estimators, like ensemble methods) within sklearn and also as additional packages (such as XGBoost). AdaBoost is one of the most popular and versatile. Its loss function is a special case of the more general gradient boosted trees.

### Steps 1-4
We follow the `sklearn` training process as seen on the slides.

We use [`sklearn.ensemble.AdaBoostClassifier`](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostClassifier.html). The key hyperparameters are:
* `base_estimator`: the base estimator (such as a decision tree) to boost
* `n_estimators`: the number of base estimators, $B$; the default is $50$, which is a good starting point; unlike bagging, boosting can overfit as $B\rightarrow\infty$; use CV to find the best $B$
* `learning_rate`: the learning rate, $\lambda$; how much each new estimator contributes to the overall estimator; $\lambda>1$ faster learning, fewer $B$; $\lambda<1$ slower learning, more $B$


In addition, we can specify the hyperparameters of the base estimator. The default base estimator is a decision tree stump, `DecisionTreeClassifier(max_depth=1)`, which is the recommended starting point. Because slow learners of weak base estimators have shown the best results, we want to keep the base estimator simple. So unlike in bagging, we do want to limit the individual trees. A common range for `max_depth` is $[1, 10)$. In some text books this is also called the *interaction depth*, because the number of splits determines the number of interactions between variables.

```python
from sklearn.ensemble import AdaBoostClassifier

# choose model & hyperparameters
model = AdaBoostClassifier()

# fit
model.fit(X_train, y_train)
```

Out of the box, the model doesn't work even half as well as the previous bagging estimator:

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix

y_fit = model.predict(X_test)
cm = confusion_matrix(y_test, y_fit)
sns.heatmap(cm.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

### Improving AdaBoost

Careful, this takes a while:

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

param_grid = {'base_estimator__max_depth': [1, 2, 4],
              'n_estimators': [50, 100, 200, 500, 1000],
              'learning_rate': [.001, .01, .1, .5, 1, 1.5]}

grid = GridSearchCV(AdaBoostClassifier(base_estimator=DecisionTreeClassifier()),
                    param_grid, cv=5)

grid.fit(X_train, y_train)

grid.best_params_
```

```python
grid.best_score_
```

```python
y_fit = grid.predict(X_test)
cm = confusion_matrix(y_test, y_fit)
sns.heatmap(cm.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```
