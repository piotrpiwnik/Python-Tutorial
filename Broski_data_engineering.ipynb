```python
#Load the data from a csv file (broski.csv) from Desktop from "C:\Users\piotr\OneDrive - Erasmus University Rotterdam\Desktop\broski.xlsx"
# Path: Python Tutorial\Broski_data_engineering.ipynb. Use 
df = pd.read_excel(r'C:\Users\piotr\OneDrive - Erasmus University Rotterdam\Desktop\broski.xlsx')
```

```python
df.head(3)
```

```python
#drop the second row of the dataframe df
df.drop(df.index[0], inplace=True)
#Change the name of the first column to "Date"
df.rename(columns={'Feature': 'Date', 'P' : 'P.0', 'MV' : 'MV.0', 'PTBV' : 'PTBV.0', 'ESG' : 'ESG.0' }, inplace=True)
```

```python
df
```

```python
# Set 'Date' as the index if not already set
df.set_index('Date', inplace=True)

# Convert the columns to a MultiIndex
df.columns = pd.MultiIndex.from_tuples(df.columns.str.split('.', expand=True))

# Unstack the dataframe
df_long = df.unstack().reset_index()

# Rename columns
df_long.columns = ['ID', 'Company_ID', 'Date', 'Value']

# Pivot to get the desired structure
df_pivot = df_long.pivot(index=['Company_ID', 'Date'], columns='ID', values='Value').reset_index()

#Change the order of the columns in df_pivot
df_pivot = df_pivot[['Company_ID', 'Date', 'P', 'MV', 'PTBV', 'ESG']]

print(df_pivot)

```
