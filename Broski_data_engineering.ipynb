```python
import pandas as pd
import numpy as np

def transform_data(df):
    # Extract the column names and split them into feature names and company IDs
    columns = df.columns
    feature_names = [col.split('.')[0] for col in columns]
    company_ids = [col.split('.')[1] if '.' in col else '1' for col in columns]

    # Drop the first two rows from df
    df = df.iloc[2:].reset_index(drop=True)

    # Get the dates from the 'Feature' column
    dates = df['Feature'].values

    # Prepare a list of dictionaries for the final DataFrame
    data = []
    for i, date in enumerate(dates):
        for j, company_id in enumerate(company_ids):
            # Prepare a row for the final DataFrame
            row = {'Company ID': company_id, 'Date': date}
            for k, feature_name in enumerate(feature_names):
                row[feature_name] = df.iloc[i, k]
            data.append(row)

    # Convert the list of dictionaries to a DataFrame
    df_final = pd.DataFrame(data)

    return df_final
```

```python
# df = pd.read_csv('your_data.csv')  # Replace 'your_data.csv' with the path to your actual CSV file
df_transformed = transform_data(df)
```
